```python
# Python imports
from IPython.display import Image
from streng.ppp.sections.geometry.tee import TeeSectionGeometry
from streng.ppp.sections.geometry.rectangular import RectangularSectionGeometry
```

### Γεωμετρία - Δομικά στοιχεία

```python
L = 5.
H = 3.

tbeam = TeeSectionGeometry(bw=0.25, h=0.70, beff=1.20, hf=0.15)
beam_I = tbeam.moment_of_inertia_xx
beam_sheararea = tbeam.shear_area_2

col = RectangularSectionGeometry(b=0.40, h=0.40)
col_I = col.moment_of_inertia_xx
col_sheararea = col.shear_area_2

print('Δοκός')
print(f'beam_I = {beam_I:.3e}m4')
print(f'beam_sheararea = {beam_sheararea:.3e}m2')

print('')
print('Στύλος')
print(f'col_I = {col_I:.3e}m4')
print(f'col_sheararea = {col_sheararea:.3e}m2')
```

### Φορτία - Μάζα

```python
g = 20.
q = 10.
p = g+0.3*q
m = p * L /9.81
print(f'p = g+0.3q = {p:.2f}kN/m')
print(f'm = {m:.3f}t')
```
