```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Bare
fname = r'D:\mypythons\jupyters\phd\one_storey_frame\drain\onest\bare\out'

with open(fname) as f:
    content = f.readlines()
    
# Infilled
fname_inf = r'D:\mypythons\jupyters\phd\one_storey_frame\drain\onest\inf\out'

with open(fname_inf) as f:
    content_inf = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
# content = [x.strip() for x in content]
```

```python
# Bare
node_xdipl_indices = [i for i, elem in enumerate(content) if 'X-DISPLACEMENTS AT NODES, TIME =' in elem]
time_values = [float(content[i][33:]) for i in node_xdipl_indices]
xdispl = [float(content[i+3][19:28]) for i in node_xdipl_indices]

# Infilled
node_xdipl_indices_inf = [i for i, elem in enumerate(content_inf) if 'X-DISPLACEMENTS AT NODES, TIME =' in elem]
time_values_inf = [float(content_inf[i][33:]) for i in node_xdipl_indices_inf]
xdispl_inf = [float(content_inf[i+3][19:28]) for i in node_xdipl_indices_inf]

f, ax = plt.subplots(figsize=(8, 5))
ax.plot(time_values, xdispl, label="bare", lw=1)
ax.plot(time_values_inf, xdispl_inf, label="infilled", lw=1)
# ax.axis([0, 0.1, 0, 500.])
ax.set_title('X-displ time-history')
ax.set_ylabel('X-displ (m)')
ax.set_xlabel('time (sec)')
ax.legend()
fig = (f, ax)
plt.show()
```

```python
# Bare
elements_results_indices = [i for i, elem in enumerate(content) if 'RESULTS FOR GROUP  2 R.C. BEAM ELEMENTS TIME =' in elem]

element_nodei_results = [content[i+6][10:].split() for i in elements_results_indices]
element_nodej_results = [content[i+8][10:].split() for i in elements_results_indices]

base_shears = np.array(element_nodei_results).transpose().astype(np.float)[3] + \
              np.array(element_nodej_results).transpose().astype(np.float)[3]


# Infilled
elements_results_indices_inf = [i for i, elem in enumerate(content_inf) if 'RESULTS FOR GROUP  2 R.C. BEAM ELEMENTS TIME =' in elem]

element_nodei_results_inf = [content_inf[i+6][10:].split() for i in elements_results_indices_inf]
element_nodej_results_inf = [content_inf[i+8][10:].split() for i in elements_results_indices_inf]

base_shears_inf = np.array(element_nodei_results_inf).transpose().astype(np.float)[3] + \
                  np.array(element_nodej_results_inf).transpose().astype(np.float)[3]

f, ax = plt.subplots(figsize=(8, 5))
ax.plot([0] + time_values, base_shears, label="bare", lw=1)
ax.plot([0] + time_values_inf, base_shears_inf, label="infilled", lw=1)
# ax.axis([0, 0.1, 0, 500.])
ax.set_title('Base Shear time-history')
ax.set_ylabel('Base Shear (kN)')
ax.set_xlabel('Time (sec)')
ax.legend()
fig = (f, ax)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 5))
ax.plot([0] + xdispl, base_shears, label="bare", lw=1)
ax.plot([0] + xdispl_inf, base_shears_inf, label="infilled", lw=1)
# ax.axis([-0.1, 0.1, -500, 500.])
ax.set_title('Base Shear vs Roof displacement')
ax.set_ylabel('Base Shear (kN)')
ax.set_xlabel('Roof displacement (m)')
ax.legend()
fig = (f, ax)
plt.show()
```
