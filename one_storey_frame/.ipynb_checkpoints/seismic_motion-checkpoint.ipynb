# Seismic Motion

```python
import os
import csv
import math
import numpy as np
import pandas as pd
from scipy import integrate

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

from streng.tools.seismic_motion import Record, Spectra
from streng.common.io.output import OutputTable
```

## Διέγερση

### Φόρτωμα επιταχυνσιογραφήματος

```python
rec = Record.load_multi_column_txt(filename='LEFL', dt=0.01, skiprows=2)
rec.calc()
```

### Γραφήματα διέγερσης vs time

```python
f, ax = plt.subplots(figsize=(14, 10))

plt.title('Διέγερση')

plt.subplot(3, 1, 1)
plt.plot(rec.time, rec.accel, '-')
plt.ylabel('Acceleration (m/s2)')

plt.subplot(3, 1, 2)
plt.plot(rec.time, rec.velocity, '-')
plt.ylabel('Velocity (m/s)')

plt.subplot(3, 1, 3)
plt.plot(rec.time, rec.displacement, '-')
plt.ylabel('displacement (m)')

plt.xlabel('time (s)')

fig = (f, ax)
plt.show()
```

## Φάσματα

```python
spec = Spectra(accel=rec.accel,
              dt=rec.dt)
```

```python
# spec.get_spectra(0.05)
```

```python
dt = pd.DataFrame(spec.get_spectra(0.05), columns=['T', 'Sa', 'Sv', 'Sd', 'PSa', 'PSv'])
dt.head(10)
```

```python
spectra005 = spec.get_spectra(damping=0.05, 
                 startT=0., 
                 endT=4., 
                 noTs=401)

spectra020 = spec.get_spectra(damping=0.20, 
                 startT=0., 
                 endT=4., 
                 noTs=401)
```

```python
f, ax = plt.subplots(figsize=(9, 6))
ax.plot(spectra005['T'], spectra005['Sa'], label='damping 5%')
ax.plot(spectra005['T'], spectra020['Sa'], label='damping 20%')
# ax.axis([0, 0.3, 0, 300.])
ax.set_title('Acceleration spectra')
ax.set_ylabel('Spectral acceleration (m/s2)')
ax.set_xlabel('Period (s)')
ax.legend()
fig = (f, ax)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(9, 6))
ax.plot(spectra005['T'], spectra005['Sv'], label='Sv')
ax.plot(spectra005['T'], spectra020['PSv'], label='PSv')
# ax.axis([0, 0.3, 0, 300.])
ax.set_title('Velocity spectra')
ax.set_ylabel('Spectral velocity (m/s)')
ax.set_xlabel('Period (s)')
ax.legend()
fig = (f, ax)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(9, 6))
ax.plot(spectra005['T'], spectra005['Sd'], label='Sd')
# ax.axis([0, 0.3, 0, 300.])
ax.set_title('Displacement spectrum')
ax.set_ylabel('Spectral displacement (m)')
ax.set_xlabel('Period (s)')
ax.legend()
fig = (f, ax)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(9, 6))
ax.plot(spectra005['Sd'], spectra005['Sa'], label='Sa-Sd')
# ax.axis([0, 0.3, 0, 300.])
ax.set_title('Sa-Sd spectrum')
ax.set_ylabel('Spectral acceleration (m/s2)')
ax.set_xlabel('Spectral displacement (m/s)')
ax.legend()
fig = (f, ax)
plt.show()
```
