```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load pushover curve from sap2000 output

```python
df = pd.read_excel('sap/01_sap_auto.xlsx', skiprows=[0,2], usecols=[0,1,2])
df.head()
```

```python
sap_d = df['Displacement'].tolist()
sap_V = df['BaseForce'].tolist()
```

```python
f, ax = plt.subplots(figsize=(10, 6))
ax.plot(sap_d, sap_V, label="sap2000", lw=2)
# ax.plot(etabs_d, etabs_V, label="etabs", lw=2)
ax.axis([0, 0.20, 0, 1.2 * max(sap_V)])
ax.set_title('Καμπύλη αντίστασης')
ax.set_ylabel('V (kN)')
ax.set_xlabel('δ (m)')
ax.legend()
fig = (f, ax)
plt.show()
```
